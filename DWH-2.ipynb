```python
hd_new_curr_recs = """
 SELECT     t.skey,
            s.EmployeeID,
            s.LastName,
            s.FirstName,
            s.Title,
            s.TitleOfCourtesy,
            s.BirthDate,
            s.HireDate,
            s.Address,
            s.City,
            s.Region,
            s.PostalCode,
            s.Country,
            s.HomePhone,
            s.Extension,
            s.Photo,
            s.Notes,
            s.ReportsTo,
            s.PhotoPath,
            s.Salary,
          DATE(CURRENT_TIMESTAMP)
              AS effectivedate,
          DATE('9999-12-31') AS expiredate,
          BOOLEAN(1) AS iscurrent
 FROM     Employees_data s
          INNER JOIN current_scd2 t
              ON t.Salary = s.Salary
              AND t.iscurrent = 1
 WHERE    
 
          NVL(s.EmployeeID, '') <> NVL(t.EmployeeID, '')
          OR NVL(s.Title, '') <> NVL(t.Title, '')
           
         
         
"""
df_new_curr_recs = spark.sql(hd_new_curr_recs)
df_new_curr_recs.createOrReplaceTempView("new_curr_recs")
# ############## review dataset ############## #
df_new_curr_recs.show(10, False)
```

```python
df_modfied_keys = df_new_curr_recs.select("skey")
df_modfied_keys.createOrReplaceTempView("modfied_keys")
```

```python
 hd_new_hist_recs = """
 SELECT
 
            t.skey,
            t.EmployeeID,
            t.LastName,
            t.FirstName,
            t.Title,
            t.TitleOfCourtesy,
            t.BirthDate,
            t.HireDate,
            t.Address,
            t.City,
            t.Region,
            t.PostalCode,
            t.Country,
            t.HomePhone,
            t.Extension,
            t.Photo,
            t.Notes,
            t.ReportsTo,
            t.PhotoPath,
            t.Salary,
            t.effectivedate,
          DATE_SUB(
              DATE(CURRENT_TIMESTAMP), 1
          ) AS expiredate,
          BOOLEAN(0) AS iscurrent
 FROM     current_scd2 t
          INNER JOIN modfied_keys k
              ON k.skey = t.skey
 WHERE    t.iscurrent = 1
 

          
"""
df_new_hist_recs = spark.sql(hd_new_hist_recs)
df_new_hist_recs.createOrReplaceTempView("new_hist_recs")
# ############## review dataset ############## #
df_new_hist_recs.show(10, False)
```

```python
 hd_unaffected_recs = """
 SELECT     s.skey,
            s.EmployeeID,
            s.LastName,
            s.FirstName,
            s.Title,
            s.TitleOfCourtesy,
            s.BirthDate,
            s.HireDate,
            s.Address,
            s.City,
            s.Region,
            s.PostalCode,
            s.Country,
            s.HomePhone,
            s.Extension,
            s.Photo,
            s.Notes,
            s.ReportsTo,
            s.PhotoPath,
            s.Salary,
            s.effectivedate,
            s.expiredate,
            s.iscurrent
 FROM     current_scd2 s
          LEFT OUTER JOIN modfied_keys k
              ON k.skey = s.skey
 WHERE    
             k.skey IS NOT NULL
"""
df_unaffected_recs = spark.sql(hd_unaffected_recs)
df_unaffected_recs.createOrReplaceTempView("unaffected_recs")
# ############## review dataset ############## #
df_unaffected_recs.show(10, False)
    
    
    
    
    
    
    
    
    
    
    
    
    
```

```python
 hd_new_emp = """
 SELECT     
            s.EmployeeID,
            s.LastName,
            s.FirstName,
            s.Title,
            s.TitleOfCourtesy,
            s.BirthDate,
            s.HireDate,
            s.Address,
            s.City,
            s.Region,
            s.PostalCode,
            s.Country,
            s.HomePhone,
            s.Extension,
            s.Photo,
            s.Notes,
            s.ReportsTo,
            s.PhotoPath,
            s.Salary,
          DATE(CURRENT_TIMESTAMP) 
              AS effectivedate,
          DATE('9999-12-31') AS expiredate,
          BOOLEAN(1) AS iscurrent
 FROM     Employees_data s
          LEFT OUTER JOIN current_scd2 t
              ON t.Region = s.Region
 WHERE    t.Region IS NULL
"""
df_new_emp = spark.sql(hd_new_emp)
df_new_emp.createOrReplaceTempView("new_emp")
# ############## review dataset ############## #
df_new_emp.show(10, False)
```

```python
 v_max_key = spark.sql(
    "SELECT STRING(MAX(skey)) FROM current_scd2"
).collect()[0][0]
hd_new_scd2_a = """
 WITH a_cte
 AS   (
       SELECT
         s.employeeID,
         s.LastName,
         s.FirstName,
         s.Title,
         s.TitleOfCourtesy,
         s.BirthDate,
         s.HireDate,
         s.Address,
         s.City,
         s.Region,
         s.PostalCode,
         s.Country,
         s.HomePhone,
         s.Extension,
         s.Photo,
         s.Notes,
         s.ReportsTo,
         s.PhotoPath,
         s.Salary,
         s.effectivedate,
         s.expiredate,
         s.iscurrent
        FROM       new_emp s
        UNION ALL
        SELECT
         y.employeeID,
         y.LastName,
         y.FirstName,
         y.Title,
         y.TitleOfCourtesy,
         y.BirthDate,
         y.HireDate,
         y.Address,
         y.City,
         y.Region,
         y.PostalCode,
         y.Country,
         y.HomePhone,
         y.Extension,
         y.Photo,
         y.Notes,
         y.ReportsTo,
         y.PhotoPath,
         y.Salary,
         y.effectivedate,
         y.expiredate,
         y.iscurrent
        FROM       new_curr_recs y
      )
SELECT ROW_NUMBER() OVER(ORDER BY a.effectivedate) + BIGINT('{v_max_key}') AS skey,
          a.employeeID,
         a.LastName,
         a.FirstName,
         a.Title,
         a.TitleOfCourtesy,
         a.BirthDate,
         a.HireDate,
         a.Address,
         a.City,
         a.Region,
         a.PostalCode,
         a.Country,
         a.HomePhone,
         a.Extension,
         a.Photo,
         a.Notes,
         a.ReportsTo,
         a.PhotoPath,
         a.Salary,
         a.effectivedate,
         a.expiredate,
         a.iscurrent
FROM a_cte a
"""
df_new_scd2_a = spark.sql(hd_new_scd2_a.replace("{v_max_key}", v_max_key)) 
# ############## review dataset ############## #
df_new_scd2_a.show(10, False) 
df_new_scd2_a.createOrReplaceTempView("df_new_scd2_a")



```

```python
hd_new_scd2_b = """SELECT skey, employeeID, LastName, FirstName, Title, TitleOfCourtesy, BirthDate, HireDate, Address, City, Region, PostalCode, Country, HomePhone, Extension, Photo, Notes, ReportsTo, PhotoPath, Salary, effectivedate, expiredate, iscurrent
FROM (
    SELECT skey, employeeID, LastName, FirstName, Title, TitleOfCourtesy, BirthDate, HireDate, Address, City, Region, PostalCode, Country, HomePhone, Extension, Photo, Notes, ReportsTo, PhotoPath, Salary, CAST(NULL AS BOOLEAN) AS iscurrent, effectivedate, expiredate
    FROM df_new_scd2_a
    UNION ALL
    SELECT skey, employeeID, LastName, FirstName, Title, TitleOfCourtesy, BirthDate, HireDate, Address, City, Region, PostalCode, Country, HomePhone, Extension, Photo, Notes, ReportsTo, PhotoPath, Salary, CAST(iscurrent AS BOOLEAN) AS iscurrent, effectivedate, expiredate
    FROM unaffected_recs
    UNION ALL
    SELECT skey, employeeID, LastName, FirstName, Title, TitleOfCourtesy, BirthDate, HireDate, Address, City, Region, PostalCode, Country, HomePhone, Extension, Photo, Notes, ReportsTo, PhotoPath, Salary, CAST(iscurrent AS BOOLEAN) AS iscurrent, effectivedate, expiredate
    FROM new_hist_recs
)"""

df_new_scd2_b = spark.sql(hd_new_scd2_b)

# review dataset
df_new_scd2_b.show(50, False)

# create temporary view
df_new_scd2_b.createOrReplaceTempView("df_new_scd2_b")
```
